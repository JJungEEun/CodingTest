```python
MAX = 100;
  
class Deque:
    def __init__(self, size):
        self.arr = [0] * MAX
        self.front = -1;
        self.rear = 0;
        self.size = size;
      
      
    ''' Operations on Deque:
    void  insertfront(int key);
    void  insertrear(int key);
    void  deletefront();
    void  deleterear();
    bool  isFull();
    bool  isEmpty();
    int  getFront();
    int  getRear(); '''
    
    # Deque가 full인지 확인하기
    def isFull(self):
        return ((self.front == 0 and self.rear == self.size-1) or self.front == self.rear + 1)
      
      
    # Deque가 empty인지 확인하기.
    def isEmpty (self):
        return (self.front == -1);
      
    # front에 데이터 삽입
    def insertfront(self, key):
  
        # 1. deque가 full인지 확인
        if (self.isFull()):
            print("Overflow");
            return;    
        # 2. deque가 완전히 비어있다면 front 0으로 초기화
        if (self.front == -1):
            self.front = 0;
            self.rear = 0;
          
           
        # 3. 초기 위치의 front -> 마지막 인덱스
        elif (self.front == 0):
            self.front = self.size - 1 ;
    
        else: #4, 그렇지 않으면 1 만큼 감소
            self.front = self.front-1;
    
        # front에 데이터 삽입
        self.arr[self.front] = key ;
      
      
    # rear에 데이터 삽입
    def insertrear(self, key):
        # 1. deque가 full인지 확인:
        if (self.isFull()):
            print(" Overflow");
            return;
          
    
        # 2. deque가 완전히 비어있다면 front 0으로 초기화
        if (self.front == -1):
            self.front = 0;
            self.rear = 0;
          
    
        # 3. rear가 마지막 인덱스라면 초기화
        elif (self.rear == self.size-1):
            self.rear = 0;
    
        #4, 그렇지 않으면 1 만큼 증가
        else:
            self.rear = self.rear+1;
           
        # rear에 데이터 삽입
        self.arr[self.rear] = key ;
      
      
    #front 데이터 삭제
    def deletefront(self):
         # 1. deque가 empty인지 확인
        if (self.isEmpty()):
            print("Queue Underflow");
            return ;
          
    
        # 2, 하나의 원소만 있을 경우
        if (self.front == self.rear):
            self.front = -1;
            self.rear = -1;
          
        else:
            # 3. 초기 포지션으로 돌아가기
            if (self.front == self.size -1):
                self.front = 0;
    
            else: # 4. 요소 삭제후 1 만큼 증가
                self.front = self.front+1;
      
      
    #rear 데이터 삭제
    def deleterear(self):
        #1. deque가 비어있는지 확인
        if (self.isEmpty()):
            print(" Underflow");
            return ;
          
    
        # 2, 하나의 원소만 있을 경우
        if (self.front == self.rear):
            self.front = -1;
            self.rear = -1;
          
        elif (self.rear == 0):
            self.rear = self.size-1;
        else:
            self.rear = self.rear-1;
      
      
    # Returns front element of Deque
    def getFront(self):
        if (self.isEmpty()):
            print(" Underflow");
            return -1 ;
          
        return self.arr[self.front];
      
      
    # function return rear element of Deque
    def getRear(self):
        if(self.isEmpty() or self.rear < 0):
            print(" Underflow");
            return -1 ;
          
        return self.arr[self.rear];
```

```python
# Driver program to test above function
dq = Deque(5);
            
print("Insert element at rear end  : 5 ");
dq.insertrear(5);
  
print("insert element at rear end : 10 ");
dq.insertrear(10);
  
print(f"get rear element : {dq.getRear()}");
  
dq.deleterear();
print(f"After delete rear element new rear become : {dq.getRear()}");
  
print("inserting element at front end");
dq.insertfront(15);
  
print(f"get front element: {dq.getFront()}");
  
dq.deletefront();
  
print(f"After delete front element new front become : {dq.getFront()}");
  
```
