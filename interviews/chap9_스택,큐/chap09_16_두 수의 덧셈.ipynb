### 두 수의 덧셈

**역순으로 저장된 연결 리스트이 숫자를 더하라**

```
(2->4->3) + (5->6->4)
(7->0->8)

------------
342 + 564
```

### Solution 01

````python
def addTwoNumbers(self, l1, l2):
        
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """        
        
        # 각 List별 합 구하기 
        value1, value2 = 0, 0
    
        num = 1
        while l1:
            value1 += l1.val*num
            l1 = l1.next
            num *= 10
                        
                
        num = 1
        while l2:
            value2 += l2.val*num
            l2 = l2.next
            num *= 10
            
        value = value1+value2 # 전체합
        
        # value 값을 listnode로 변환      
        
        head = None
        for n in str(value):
            node = ListNode(n)
            node.next = head
            head = node
            
        return node
````

### Solution 02

````python
def addTwoNumbers(self, l1, l2):    

    """
    :type l1: ListNode
    :type l2: ListNode
    :rtype: ListNode
    """

    result = root = ListNode(0)
    carry = 0

    while l1 or l2 or carry:            
        val1  = (l1.val if l1 else 0)
        val2  = (l2.val if l2 else 0)
        
        # 몫, 나머지 계산
        carry, value = divmod(val1+val2 + carry, 10)    

        root.next = ListNode(value)
        root = root.next                      

        l1 = (l1.next if l1 else None)
        l2 = (l2.next if l2 else None)

    return result.next
````
